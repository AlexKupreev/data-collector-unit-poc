```python
import pandas as pd
```

```python
isd_stations_file = "../../../data/isd-history-clean.txt"
isd_stations = pd.read_fwf(isd_stations_file)
isd_stations.head()
```

```python
isd_stations["end_date"] = pd.to_datetime(isd_stations["END"], format="%Y%m%d")
isd_stations.head()
```

```python
isd_stations_icao_actual = isd_stations[(~isd_stations["CALL"].isna()) & (isd_stations["end_date"] > pd.Timestamp(year=2025, month=1, day=1))]
isd_stations_icao_actual.head()
```

```python
import gzip

colspecs = [
    (0, 4),    # Year
    (5, 7),    # Month
    (8, 10),   # Day
    (11, 13),  # Hour
    (14, 19),  # Air Temperature
    (20, 25),  # Dew Point Temperature
    (26, 31),  # Sea Level Pressure
    (32, 37),  # Wind Direction
    (38, 43),  # Wind Speed Rate
    (44, 49),  # Sky Condition
    (50, 55),  # Precipitation Depth 1 hour
    (56, 61),  # Precipitation Depth 6 hours
]

# Decompress the gzipped content
with gzip.open("../../../data/010460-99999-2025.gz", 'rt') as gz_file:
    df = pd.read_fwf(
        gz_file,
        colspecs=colspecs,
        header=None,
        na_values=-9999,
        dtype="Int64"
    )
    
df.head(20)
```
